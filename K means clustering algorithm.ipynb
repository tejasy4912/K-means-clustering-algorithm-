**1. Importing the libraries**

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

**2. Importing the dataset**

```python
df = pd.read_csv('Housing.csv')
df.head()
```

**3. Visualizing the data**

```python
plt.scatter(df['price'],df['area'])
plt.show()
```

**4. Selecting the features**

```python
data = df.loc[:, ["price", "area"]]
print(data.head())
```

**5. Applying K-means**

```python
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
identified_clusters = kmeans.fit_predict(data)
print(identified_clusters)
```

**5. Visualizing the clusters formed**

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['price'],data_with_clusters['area'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
